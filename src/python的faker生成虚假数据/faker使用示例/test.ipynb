```python
# ! 随机性别
class Gender:
    def __init__(self, lang="中文"):
        """默认为中文"""
        if lang == "英语":
            self.lang = "英文"
        else:
            self.lang = "中文"

    def create(self, n=10):
        if self.lang == "中文":
            for _ in range(n):
                r = random.randint(0, 1)
                if r == 0:
                    print('男')
                else:
                    print("女")
        elif self.lang == "英文":
            for _ in range(n):
                r = random.randint(0, 1)
                if r == 0:
                    print('male')
                else:
                    print("female")
```

```python
gender = Gender()
gender.create()
```

```python
# ! 随机时间
from time import mktime,localtime,strftime
class Time:
    def __init__(self, distinct="zh_CN"):
        self.distinct = distinct
        self.fake = Faker(locale=self.distinct)

    def create(self):
        a1 = (2015, 4, 12, 0, 0, 0, 0, 0, 0)  # 设置开始日期时间元组（2020-04-12 00：00：00）
        a2 = (2020, 4, 13, 0, 0, 0, 0, 0, 0)  # 设置结束日期时间元组（2020-04-13 00：00：00）

        start = mktime(a1)  # 生成开始时间戳
        print("start时间戳:", start)
        end = mktime(a2)  # 生成结束时间戳
        print("end时间戳:", end)

        # 随机生成10个日期字符串
        for i in range(10):
            t = random.randint(start, end)  # 在开始和结束时间戳中随机取出一个
            date_touple = localtime(t)  # 将时间戳生成时间元组
            date_str = strftime("%Y-%m-%d %H:%M:%S", date_touple)  # 将时间元组转成格式化字符串（1976-05-21）
            print(date_str)

    def fakeCreate(self, n=10):
        for _ in range(n):
            print(self.fake.date())
```

```python
time = Time()
time.create()
```

```python
class Proxy:
    def __init__(self,distinct="zh_CN"):
        self.distinct = distinct
        self.fake = Faker(locale=self.distinct)

    def create(self):
        print(self.fake.android_platform_token())  # 安卓
        print(self.fake.chrome(version_from=13, version_to=63, build_from=800, build_to=899))  # Chrome
        print(self.fake.firefox())  # FireFox
        print(self.fake.internet_explorer())   # Ie
        print(self.fake.ios_platform_token())  # ios
        print(self.fake.linux_platform_token())  # Linux
        print(self.fake.linux_processor())   # Linux处理器
        print(self.fake.mac_platform_token()) # Mac
        print(self.fake.mac_processor())    # Mac处理器
        print(self.fake.opera())   # Opera
        print(self.fake.safari())  # Safari
        print(self.fake.user_agent())   # 随机用户代理
        print(self.fake.windows_platform_token())  # Windows
```

```python
proxy = Proxy()
proxy.create()
```
